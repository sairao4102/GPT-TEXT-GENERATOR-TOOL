 # Text Generation with GPT-2 Medium 

This notebook demonstrates how to use the gpt2-medium model from Hugging Face's transformers library to generate human-like text based on a prompt. GPT-2 Medium is a powerful language model trained to predict the next word in a sequence.

### -->Load the Pre-trained GPT-2 Medium Model

```python
from transformers import AutoTokenizer, GPT2LMHeadModel, pipeline

# Load GPT2-Medium tokenizer and model
model_name = "gpt2-medium"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

# Save the model locally (use consistent naming)
model.save_pretrained("model/gpt2-medium")
tokenizer.save_pretrained("model/gpt2-medium")

# Create a text generation pipeline
generator = pipeline("text-generation", model=model, tokenizer=tokenizer, temperature=0.9)
```

### -->Define the Text Generation Function

```python
def generate_text(prompt):
    result = generator(prompt, max_length=250, num_return_sequences=1)
    return result[0]['generated_text']
```
